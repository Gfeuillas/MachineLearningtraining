```python
import pandas as pd #ok
from sklearn.cluster import KMeans #~+
from sklearn import datasets #ok
import numpy as np #ok
import matplotlib.pyplot as plt #ok
from mpl_toolkits.mplot3d import Axes3D #~-
```

```python

iris=datasets.load_iris()#ok
dfi=pd.DataFrame(iris.data, columns=iris.feature_names)#~+
dfi['target']=iris.target# ok
dfi
```

```python
km3=KMeans(3)#ok

km3.fit(dfi.iloc[:,[0,1,2,3]])#ok
```

```python
fig= plt.figure(1, figsize=(9,10))# ~~
ax= Axes3D(fig,rect=[0,0,.95,1]) # ~~ 
ax.scatter(dfi.iloc[:,3],dfi.iloc[:,0],dfi.iloc[:,2],# ~+
          c=km3.labels_, edgecolor="k")
```
